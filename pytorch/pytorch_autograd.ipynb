```python
import torch
from torch import FloatTensor
from torch.autograd import Variable
from torchviz import make_dot

import numpy as np
```

```python
x = Variable(FloatTensor([4]),requires_grad=True)
y = x ** 2

#do backward propagation
y.backward()

# The gradient of the power function at x = 4 equals 8 (2 * 4)
assert float(x.grad) == 8
```

```python
x = Variable(FloatTensor([4]),requires_grad=True)
y = x ** 2
y.retain_grad()
z = 5 * y
z.backward()

assert float(y.grad) == 5
assert float(x.grad) == 40

make_dot(z,{'x' : x})
```

```python
x = torch.ones(2, 2, requires_grad=True)
y = x + 2
z = y * y * 3
out = z.mean()
out.backward()
```

```python
print(x.grad)
#https://pytorch.org/tutorials/beginner/blitz/autograd_tutorial.html
```

```python
a = Variable(FloatTensor([4]))

weights = [Variable(FloatTensor([i]), requires_grad=True) for i in (2, 5, 9, 7)]

# unpack the weights for nicer assignment
w1, w2, w3, w4 = weights

b = w1 * a
c = w2 * a
d = w3 * b + w4 * c
L = (10 - d)

L.backward()
```

```python
make_dot(L,{'a':a,'w1':w1,'w2':w2,'w3':w3,'w4':w4,'b':b,'c':c,'d':d})
```
